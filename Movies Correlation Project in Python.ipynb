```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')

from matplotlib.pyplot import figure
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
df = pd.read_csv('movies.csv')
```

```python
df.head()
```

```python
#When you want to see all the rows in the dataframe
pd.set_option('display.max_rows', None)
```

```python
df.head()
```

```python
#Checking for null values in columns.

for col in df.columns:
    pct_missing = np.mean(df[col].isnull()) #compute the mean of Boolean mask (True evaluates as 1 and False as 0)
    print(f'{col} - {pct_missing}%')
```

```python
#checking data types in dataframe
df.dtypes
```

```python
df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
#Create a new Column

df['correct_year'] = df['released'].astype('str').str[:4]
```

```python
df.head()
```

```python
df.head()
```

```python
#Order by gross revenue

df2 = df.sort_values(by=['gross'], inplace = False, ascending = False)
```

```python
#Drop any duplicates
#show us what values are distinctive in company value

df2['company'].drop_duplicates().sort_values(ascending = False)


```

```python
df2.head()
```

```python
#Project Assumptions
#budget high correlation
#company high correlation
```

```python
# scatter plot to compare gross vs budget

plt.scatter(x= df2['budget'], y = df2['gross'])

plt.title('Budget Vs Gross')
plt.xlabel('Budget for film')
plt.ylabel('Gross Earnings')

plt.show
```

```python
# Plot Budget vs gross using seaborn

sns.regplot(x = 'budget', y = 'gross', data = df2, scatter_kws={'color':'red'},line_kws = {'color':'blue'})
```

```python
df2.head()
```

```python
# Now let's have a look at correlation
```

```python
df2.corr(method = 'pearson') #pearson, kendall, spearman
```

```python
#Assumption was correct, budget and gross have high correlation
```

```python
correlation_matrix = df2.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation for numeric features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()

#Anything is black is low correlation and anything is in lighter colors gives more correlation.
```

```python
#Converting Object type to category numeric type

df_numerized = df2

for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
#Comparing df_numerized with our orginial df
```

```python
df.sort_values(by=['gross'], inplace = False, ascending = False).head()
```

```python
correlation_matrix = df_numerized.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation for numeric features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```
